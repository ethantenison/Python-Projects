```python
# import sys # rarely need this

#import pyreadr as Rr
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt   #99% of time this is all you need from matplotlib
import seaborn as sns  # improves graphics, some useful new prefab charts
# import plotnine as gg    #ggplot2 ported to Python/Pandas for users/cult members
# needs to be installed using "conda install"

%matplotlib inline      
#display plots in Jupyter
print(plt.style.available)
plt.style.use('fivethirtyeight')

```

```python
np.random.seed(4718952)
# increase nsmpl to demonstrate unbiasedness
# increase nobs to demonstrate consistency
nobs=30
nsmpl=50
```

```python
x1=np.random.randint(0,20,(nsmpl,nobs)) 
x1.shape
```

```python
yn=20+.4*x1+np.random.randn(nobs)
yp=19+.4*x1+np.random.poisson(1,nobs)
```

```python
import statsmodels.formula.api as smf
betan=[]
betap=[]
for i in range(0,nsmpl):
    dfi=pd.DataFrame({'x':x1[i,:],'yn':yn[i,:],'yp':yp[i,:]})
    modn=smf.ols('yn~x',dfi).fit()
    modp=smf.ols('yp~x',dfi).fit()
    betan.append(modn.params[1])
    betap.append(modp.params[1])
bs=pd.DataFrame({'bn':betan,'bp':betap})
print(bs.describe())
fig,ax=plt.subplots(figsize=(12,9))
ax.hist(bs.bn,bins=100,alpha=.2,color='b')
ax.hist(bs.bp,bins=100,alpha=.2,color='r')
ax.legend(['normal disturbance','poisson disturbance']);
```
